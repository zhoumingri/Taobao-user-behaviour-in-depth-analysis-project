```python
用户喜好分析
1. PV/PV 转化率分析
基于行为次数的转化率，分析用户在每个环节的行为发生频率以及整个购物流程的效率。
即行为转化效率
比如，每100次商品页面浏览（PV），有多少次会转化为收藏或加购行为（PV），以及最终有多少次会转化为支付购买行为（PV）。
说明数据集中' behavior_type' 列数值含义：
1 ：浏览
2 ：收藏
3： 加购 
4： 购买

从以下三种路径计算行为转化效率

路径1：浏览 → 购买  衡量从“浏览”到“购买”的转化效率。

路径2：浏览 → 收藏 +加购  衡量从“浏览”到“产生兴趣”的转化效率。

路径3：（收藏 + 加购）→ 购买  衡量从“产生兴趣”到“购买”的转化效率。
```

```python

# 1. 统计各行为类型的 PV（访问量）
# -----------------------------
pv = df['behavior_type'].value_counts().sort_index()
browse_count = pv.get(1, 0)  # 浏览
collect_count = pv.get(2, 0) # 收藏
cart_count = pv.get(3, 0)    # 加购
buy_count = pv.get(4, 0)     # 购买

print("各行为 PV 统计：")
print(f"浏览: {browse_count}")
print(f"收藏: {collect_count}")
print(f"加购: {cart_count}")
print(f"购买: {buy_count}")


# 2. 计算三种转化率

# 路径1：整体转化率 —— 浏览 → 购买
conversion1 = (buy_count / browse_count * 100) if browse_count > 0 else 0

# 路径2：浏览 → 收藏 /加购   在所有“表达过购买意向”的用户中，有多少最终完成了购买   所有“收藏”的次数 + 所有“加购”的次数 = 总的“意向行为”次数
#然后看这些“意向行为”中有多少最终转化成了“购买”  

# 收藏 /加购 行为总次数（PV）
fav_cart_count= df[df['behavior_type'].isin([2, 3])].shape[0]

# 转化率
conversion2= (fav_cart_count/ browse_count) * 100 if browse_count > 0 else 0


# 路径3：（收藏 + 加购）→ 购买
combined_count = collect_count + cart_count  # 表达购买意向的行为总次数（收藏或加购）  是 “或”行为的总频次
conversion3 = (buy_count / combined_count * 100) if combined_count > 0 else 0

# 保存结果
conversion_rates = {
    '整体转化率\n(浏览 → 购买)': conversion1,
    '浏览→收藏/加购\n转化率': conversion2,
    '（收藏+加购）→购买\n转化率': conversion3
}

print("\n转化率结果：")
for k, v in conversion_rates.items():
    print(f"{k}: {v:.2f}%")


# 4. 可视化：横向柱状图
# -----------------------------
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# 转换为 DataFrame
conversion_df = pd.DataFrame(list(conversion_rates.items()), columns=['Conversion Path', 'Rate (%)'])

# 绘图
plt.rcParams['font.sans-serif']=['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False  #用来正常显示负号
plt.rcParams['figure.dpi']=100
ax = sns.barplot(y='Conversion Path', x='Rate (%)', data=conversion_df, palette='Blues_r')

# 添加数值标签
for i, v in enumerate(conversion_df['Rate (%)']):
    ax.text(v + 0.5, i, f"{v:.2f}%", color='black', va='center', fontsize=10)

plt.title('用户行为转化率分析', fontsize=16, fontweight='bold')
plt.xlim(0, max(conversion_df['Rate (%)']) * 1.2)  # 留白
plt.xlabel('转化率 (%)')
plt.ylabel('转化路径')
plt.tight_layout()
plt.show()

```

```python
(1)从“浏览”到“购买”的转化效率仅1.04%，
从“浏览”到“产生兴趣”的转化效率为5.07%;
(2)从“产生兴趣”到“购买”的转化效率为20.51%，说明客户浏览到感兴趣的商品成交概率较大。
```

```python
2.ARPPU分析
ARPPU(Average Revenue Per Paying User)表示每付费用户平均收益，计算公式:ARPPU=总收入/付费用户数。
该数据集中没有收益金额，这里我们改为度量每付费用户平均消费次数，计算公式:ARPPU=总消费次数/付费用户数 。
```

```python
#统计用户每日的消费次数
# 统计用户每日的消费次数
user_arppu = df[df['behavior_type'] == 4].groupby(['date', 'user_id']).size().reset_index(name='buy_count')

# 计算ARPPU
arppu = user_arppu.groupby('date').apply(lambda x: x['buy_count'].sum() / len(x)).reset_index(name='arppu')

# 可视化 - 使用 histplot
plt.figure(figsize=(10, 6))
sns.histplot(arppu['arppu'], kde=True, bins=30)
plt.title('ARPPU Distribution')
plt.xlabel('ARPPU')
plt.ylabel('Density')
plt.show()
```

```python
# 计算用户的购买频次，date得去重，因为一天多次的购买不算复购
user_pay = df[df['behavior_type'] == 4].groupby('user_id')['date'].nunique()

# 计算复购率，只取>1产生了复购的数据
repeat_buy_ratio = (user_pay[user_pay > 1].count() / user_pay.count())

print(f"复购率: {repeat_buy_ratio:.2%}")
```

```python
4.复购周期分析
了解用户多久复购一次
```

```python
#统计不同时间每个用户的购买次数
user_buy = df[df['behavior_type'] == 4].groupby(['user_id', 'date']).size().reset_index(name='purchase_count')

# 对时间进行先后排序，计算时间间隔，并去掉每个用户第一次购买记录
user_buy_date_diff = user_buy.groupby('user_id')['date'].apply(lambda x: x.sort_values().diff()[1:])
user_buy_date_diff.head()
```

```python
# 可视化
user_buy_date_diff = user_buy_date_diff.apply(lambda x: x.days)
user_buy_date_diff.value_counts().sort_index().plot(kind='line')
plt.xlabel('repeat_day_diff')
plt.ylabel('count')
plt.title('Distribution of Repurchase Cycles')
plt.show()
```

```python
#不同用户平均复购时间分布
sns.histplot(user_buy_date_diff.reset_index().groupby('user_id').date.mean(), kde=True, bins=30)
```

```python
(1)大部分用户的复购行为发生在5天之内，在第5天复购的行为出现了明显的拐点;
(2)不同用户平均复购时间呈正态分布，但是总体来看，呈现逐渐下降趋势,
多数淘宝用户平均复购时间集中在1-5天时间间隔内。
```
